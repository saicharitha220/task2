```python
import warnings
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
data = {
    'Hours': [2.5, 5.1, 3.2, 8.5, 3.5, 1.5, 9.2, 5.5, 8.3, 2.7, 7.7, 5.9, 4.5, 3.3, 1.1, 8.9, 2.5, 1.9, 6.1, 7.4, 2.7, 4.8, 3.8, 6.9, 7.8],
    'Scores': [21, 47, 27, 75, 30, 20, 88, 60, 81, 25, 85, 62, 41, 42, 17, 95, 30, 24, 67, 69, 30, 54, 35, 76, 86]
}

# Create DataFrame
df = pd.DataFrame(data)

# Define the independent variable (Hours) and dependent variable (Scores)
X = df[['Hours']]
y = df['Scores']

# Create a linear regression model
model = LinearRegression()
model.fit(X,y)

# Predict the score for a student who studies 9.25 hours/day
hours_studied = np.array([[9.25]])
predicted_score = model.predict(hours_studied)
warnings.filterwarnings('ignore',category=UserWarning)

# Display the predicted score
print(f"Predicted score for a student who studies 9.25 hours/day: {predicted_score[0]:.2f}")

# Plot the regression line
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red')
plt.title('Hours vs Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Score')
plt.show()
```
